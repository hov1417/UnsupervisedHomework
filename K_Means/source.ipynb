```python
from k_means import KMeans, KMeansPlusPlus
import numpy as np
import argparse
import matplotlib.pyplot as plt
```

```python
dataFile = "data.npy"
algorithm = KMeansPlusPlus
k = 5
name = "hov"
```

```python
data = np.load(dataFile)
kmeans = algorithm(k)
kmeans.fit(data)
labels, means = kmeans.predict(data)
colors = 'bgrcmyk'

clusters = {i: [] for i in range(k)}

for i, l in enumerate(labels):
    clusters[l].append(i)

for i in range(k):
    plt.scatter(data[clusters[i]][:, 0], data[clusters[i]][:, 1],
                c=colors[i], alpha=0.5)
plt.axis('off')
plt.savefig('K_Means_results_{}'.format(name))
plt.show()
```
